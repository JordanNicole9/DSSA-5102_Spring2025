```python
# How to connect to our .db file using library SQLAlchemy
# Notice that we are not specifiying a host, password, or server since this .db file is stored locally.
db_path = r"C:\Users\jorda\Documents\Data_Gathering\Assignment4\sql-murder-mystery.db"   
# Creating the database engine
engine = create_engine(f"sqlite:///{db_path}")

# Verifying the engine was actually created
print("Database engine created:", engine)

# Testing the connection
try: # Testing
    with engine.connect() as conn: # Establishing connection
        result = conn.execute("SELECT name FROM sqlite_master WHERE type='table';") # Selecting the name from the sqlite master (tables)
        tables = result.fetchall() # Putting the results into this 
        print("Tables in the database:", [table[0] for table in tables]) # Printing the results
except Exception as e:
    print("Error connecting to the database:", e) # Printing the results 
```

All we know at the start of our investigation is the <b>murder</b> occurred on <b>Jan.15, 2018</b> and that it took place in <b>SQL City<b>.

```python
# Write a query in SQL to view the crime scene report 
# for the murder occurred on Jan.15, 2018 and that it took place in SQL City.
# Select ALL records FROM the table WHERE...

with engine.connect() as connection:  # Establish a connection
    query = text("""SELECT * 
                    FROM crime_scene_report 
                    WHERE date = '20180115' 
                    AND city = 'SQL City' 
                    AND type = 'murder'
                    """)  # Retrieve all records matching the given date, city, and crime type
    crime_report = pd.read_sql(query, connection)  # Use pandas to read the SQL query with the connection

# Print the results
crime_report
```

Another clue! It looks like we have two witnesses! Now who are they and how do we find them?

```python
# Write two queries in SQL to find each witness in the person table. 
# For the first query as first_witness, select all the records from person where the address_street_name is Northwestern Drive and order by adress_number
# Fetch the first_witness

with engine.connect() as connection:  # Establish a connection
    query = text(
        """SELECT * 
           FROM person 
           WHERE address_street_name = 'Northwestern Dr' 
           ORDER BY address_number
        """
    )  # Retrieve all records from person table matching the given street and order by address number

    first_witness = pd.read_sql(query, connection)  # Store results in first_witness

# Print the results
first_witness
```

We have identified our first witness as ___Kinsey Erickson___. Now lets find the second!

```python
# For the third query as second_witness,
# Select all the records from person where the address_street_name is Frnaklin Ave and the name includes Annabel. 
# Fetch the second_witness

with engine.connect() as connection:  # Establish a connection
    query = text(
        """SELECT * 
           FROM person 
           WHERE address_street_name = 'Franklin Ave' 
           AND name LIKE '%Annabel%'
        """
    )  # Retrieve all records where name includes 'Annabel' and street is Franklin Ave

    second_witness = pd.read_sql(query, connection)  # Store results in second_witness

# Print the results
second_witness
```

We have identified our second witness as ___Annaebel Miller____. What were their witness statements? What did they see?

```python
# For our fourth query, witness_statements, lets find out what each witness saw...or claim they saw?
# Select the peron's name and interview where the person_id is X, X (Get the ID from the above queries)

# Replace X, X with the actual person_id of Kinsey Erickson and Annabel Miller
with engine.connect() as connection:  # Establish a connection
    query = text(
        """SELECT person.name, interview.transcript 
           FROM interview 
           JOIN person ON interview.person_id = person.id
           WHERE person.id IN (89906, 16371)  -- Replace with actual witness IDs
        """
    ) # Retrieves persons' name and interview from the person_id is X, X
    witness_statements = pd.read_sql(query, connection)  # Store results

# Print the results
witness_statements
```

```python
# Create a query to select for the following clues: (You can create multiple queries or combine them into one and go from there.)
# Remember to write our your query in the comments!

# Query to find the suspect's details based on the witness statement.
# Annabel Miller mentioned that she recognized the murderer, and the name given was "Jeremy Bowers."

with engine.connect() as connection: # Establishing a connection 
    query = text(
        """SELECT * 
           FROM person 
           WHERE name = 'Jeremy Bowers' """  # Suspect name from the witness statement
    )
    suspect_info = pd.read_sql(query, connection) # Stores results

# Print the results
suspect_info
```

We have a suspect! Lets check their witness statement. Do they have an alibi?

```python
# Query to check if the suspect has an alibi by retrieving their interview statement.
# If they were questioned, we can compare their response to the witness statement.

with engine.connect() as connection: # Establishing a connection 
    query = text(
        """SELECT * 
           FROM interview 
           WHERE person_id = (SELECT id FROM person WHERE name = 'Jeremy Bowers') """
    ) # Getting the person_id from the interview statements 
    suspect_interview = pd.read_sql(query, connection) # Storing results 

# Print the results
suspect_interview
```

Uh oh...turns out it wasn't just one person behind the crime...who is this additional person involved?

```python
# Remember to write our your query in the comments!
# You can complete this in a single query or multiple queries.

# Query to find the wealthiest individuals.
# Since there is no 'gender' column, we retrieve all high-income individuals.

with engine.connect() as connection: # Establishing a connection
    query = text(
        """SELECT person.id, person.name, income.annual_income 
           FROM person 
           JOIN income ON person.ssn = income.ssn 
           ORDER BY income.annual_income DESC
        """
    ) # Selecting the items above from the person category
    wealthy_people = pd.read_sql(query, connection) # Storing the results 

# Print the results
wealthy_people
```

Three different people match this description! Now which one attended the SQL Symphony Concert three times? 

```python
# Let's find the killer!
# Remember to write our your query in the comments!
# Query to find the person who attended the SQL Symphony Concert three times.

with engine.connect() as connection: # Establishing a connection
    query = text(
        """SELECT person_id, COUNT(*) AS concert_attendance 
           FROM facebook_event_checkin 
           WHERE event_name = 'SQL Symphony Concert' 
           GROUP BY person_id 
           HAVING COUNT(*) = 3
        """
    ) # Selecting the above items from the facebook_event_checkin, from the SQL event, and grouping byt person_id where thery attended 3 timws
    concert_attendee = pd.read_sql(query, connection) # Storing the results

# Print the results
concert_attendee
```

```python
# Query to find the wealthy individual who attended the SQL Symphony Concert three times.

with engine.connect() as connection: # Establishing connection
    query = text(
        """SELECT person.id, person.name, income.annual_income 
           FROM person 
           JOIN income ON person.ssn = income.ssn 
           WHERE person.id IN (
               SELECT person_id 
               FROM facebook_event_checkin 
               WHERE event_name = 'SQL Symphony Concert' 
               GROUP BY person_id 
               HAVING COUNT(*) = 3
           )
           ORDER BY income.annual_income DESC
        """
    ) # Selecting the wealthy individuals who attendded the SQL concert three times
    final_suspect = pd.read_sql(query, connection) # Storing the results

# Print the results
final_suspect
```
